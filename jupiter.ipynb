# Task 1

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from jupiter import Moons
```

### Exploring the dataset

```python
# Initialise the moons class
data = 'jupiter.db'
moons = Moons(data)
```

```python
conn = "sqlite:///jupiter.db"
moons_table = "SELECT * FROM moons"
moons_df = pd.read_sql(moons_table, conn)
moons_df
```

### Summary statistics

```python
sum_stats = moons.summary_statistics()
sum_stats  # prints the general statistics
```

### Correlations between variables

```python
corr = moons.correlations()
corr
```

High positive or negative correlation values indicate a strong relationship between variables.<br><br>
In this case,<br> 
- **period_days and distance_km, period_days and inclination_deg, inclination_deg and distance_km, radius_km and mass_kg** have high positive correlation values close to 1, indicating that as one increases, the other tends to increase similarly as well.
- **radius_km and mag** has the highest and strongest negative correlation, suggesting that as one increases, the other tends to decrease similarly.

### Plots

#### period_days vs distance_km

```python
pdays_vs_dist = moons.plots('period_days', 'distance_km')
```

* This scatter plot suggests a proportional relationship between the orbital period of moons (in days) and their distance from Jupiter, in which they increase proportionally.

#### period_days vs inclination_deg

```python
pdays_vs_deg = moons.plots('period_days', 'inclination_deg')
```

* This scatter plot shows that there are varying inclinations within a range of orbital periods. The clearest relationship here is that there is a large cluster of moons with higher inclinations that also have longer orbital periods. 

#### radius_km vs mass_kg

```python
rad_vs_mass = moons.plots('radius_km', 'mass_kg')
```

* This scatter plot shows an exponential relationship between radius and mass: as the radius of the moons increases, mass also tends to increase, though at a faster rate than the radius.

#### radius_km vs mag

```python
rad_vs_mag = moons.plots('radius_km', 'mag')
```

* This scatter plot shows an inverse relationship between the radius of the moons and their magnitude. It suggests that larger moons have lower magnitudes, and vice versa.
